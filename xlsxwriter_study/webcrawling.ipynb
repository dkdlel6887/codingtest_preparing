# BeautifulSoup
 - str 타입의 html 데이터를 html 구조를 가진 데이터로 가공해주는 라이브러리
 - BeautifulSoup(markup, "html.parser")
 - BeautifulSoup(markup, "lxml")
 - BeautifulSoup(markup, "lxml-xml") BeautifulSoup(markup, "xml")
 - BeautifulSoup(markup, "html5lib")

```python
from bs4 import BeautifulSoup

response = requests.get('http://www.paullab.co.kr/stock.html')
html = response.text
soup = BeautifulSoup(html, 'html.parser')
```

```python
soup.title  # tag에 바로 접근할 수 있다
```

```python
soup.title.string
```

```python
soup.title.text
```

```python
soup.title.parent.name  # tag name 출력
```

```python
soup.tr  # table row
```

```python
soup.td   # table data : 처음 만나는 테이블 데이터
```

```python
soup.th   # table head
```

```python
soup.table
```

```python
soup.find('table')
```

```python
soup.find(id=('update')).text
```

```python
soup('h2', id = '제주코딩베이스캠프연구원')
```

### Selector
 - 태그에 세밀한 접근 가능
 - class 지칭할 때는 '.', id 지칭할 때는 '#' 사용
 - 탐색하려는 태그가 특정태그 하위에 있을 경우 '>' 사용
 - Xpath full 을 사용하는게 중복없이 정확하게 찾기 좋음<br>

```python
soup.select('.tables > table > tr > th ') 
```

```python
soup.select('.main > h2')
```

```python
soup.select('.main > #제주코딩베이스캠프학원')
```
