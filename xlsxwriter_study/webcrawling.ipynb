### Quiz
 - 그룹사별 거래량 그래프로 그리기(레전드 포함), 전체그룹사 총량을 그래프로 표시

```python
import bs4
import requests
from bs4 import BeautifulSoup

response = requests.get('http://www.paullab.co.kr/stock.html')
html = response.text
soup = BeautifulSoup(html, 'html.parser')
```

```python
import re

soup.select('div.main')[2]   # 제주코딩베이스캠프 연구원
soup.select('div.main')[3]   # 제주코딩베이스캠프 공업
soup.select('div.main')[4]   # 제주코딩베이스캠프 출판사
soup.select('div.main')[5]   # 제주코딩베이스캠프 학원

datas = []
dates = []
for i in soup.select('div.main')[2:6]:
    data = []
    date = []
    for j in i.select('table>tbody>tr>td>span')[6:len(i.select('table>tbody>tr>td>span')):7]:
        j = str(j)
        j = int(re.sub('[^0-9]', '', j))
#       j = j.text.replace("/<span>",'')
        print(j)
        data.append(j)
        print(j)
    datas.append(data)  # 각 그룹사별 거래량데이터 저장
    print('=============1')
    for j in i.select('table>tbody>tr>td>span')[0:len(i.select('table>tbody>tr>td>span')):7]:
        j = str(j)
        j = re.sub('[^0-9.]', '', j)
        date.append(j)
```

```python
print(datas)
```

```python
print(date)
```

```python
import pandas as pd

data_dics = {'연구원거래량':datas[0],
             '공업거래량':datas[1],
             '출판사거래량':datas[2],
             '학원거래량':datas[3]
            }

df = pd.DataFrame(data_dics,index = date)
df
```

```python
df = df.iloc[::-1]
df
```

```python
from matplotlib import font_manager, rc

font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
rc('font', family=font_name)
lines = df.plot.line()
```

```python
df2 = df
df2['총 거래량'] = df["연구원거래량"] + df["공업거래량"] + df["출판사거래량"] + df["학원거래량"]
df2
```

```python
df2.plot(y=['총 거래량'], ylim=(0, 4000000), figsize=(10, 6))
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

fig, ax = plt.subplots()
ax.plot(df2.index,df2['총 거래량'],label = 'ALL')
ax.legend()

plt.xticks(rotation=-45)
ax.yaxis.set_major_locator(ticker.MultipleLocator(1000000))
current_values = plt.gca().get_yticks()
plt.gca().set_yticklabels(['{:.0f}'.format(x) for x in current_values])
plt.show()
```

```python
plt.plot(df.index, df["연구원거래량"],label='A')
plt.plot(df.index, df["공업거래량"],label='B')
plt.plot(df.index, df["출판사거래량"],label='C')
plt.plot(df.index, df["학원거래량"],label='D')

plt.xticks(rotation=-45)
current_values = plt.gca().get_yticks()
plt.gca().set_yticklabels(['{:.0f}'.format(x) for x in current_values])
plt.legend()
plt.show()
```
