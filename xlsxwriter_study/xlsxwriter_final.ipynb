```python
path = ('./stream_data/*.hwp')
totaldata = []

for fname in glob.glob(path):
    with ole.open(fname) as f:
        print(f.list_streams())
        totaldata.append(f.get_stream('PrvText').read().decode('utf-16le').split('><'))
        # 한글파일을 한번씩 열어서 저장해줘야 PrvText가 생김
totaldata
```

```python
def mk_df(data):
    # 데이터 두개로 나누기
    for i in range(len(data)):
        if data[i] == '연 장':
            idx = i
    data = data[idx+1:]

    # 불필요 데이터 삭제
    del data[-1]

    for i in range(0,len(data),7):
        if i == 0:
            data[i] = data[i].replace("%>\r\n<","")
        else:
            data[i] = data[i].replace(">\r\n<","")
    
    # 세부 데이터 추출
    stream_list = [
                    '계',
                    '보강천 (지방하천)',
                    '문암천 (지방하천)',
                    '문방천 (지방하천)',
                    '삼기천 (지방하천)',
                    '소하천'
                  ]
    s_detail = []
    for i in range(len(data)):
        for j in range(len(stream_list)):
            if data[i] == stream_list[j]:
                s_detail.append(data[i+1:i+7])
    
    # 데이터프레임 생성
    df = pd.DataFrame(s_detail,index = stream_list)
    df_col = [
        '연장',
        '요개수 연장',
        '개수[연장]',
        '개수[%]',
        '미개수[연장]',
        '미개수[%]',
    ]
    df.columns = df_col
    df = df.T
    
    return df
```

```python
mk_df(totaldata[0])
```

```python
for i in range(len(totaldata)-1):
    if i == 0:
        data_df = pd.concat([mk_df(totaldata[i]),mk_df(totaldata[i+1])],axis=0)
    else:
        data_df = pd.concat([data_df,mk_df(totaldata[i+1])], axis=0)
```
