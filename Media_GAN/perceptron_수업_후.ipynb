* **AND Gate weight**: 0.5, 0.5, -0.7
* **NAND Gate weight**: -0.4, -0.6, -0.6
* **OR Gate weight**: 0.5, 0.5, -0.2

```python
def AND_gate(x1, x2):
  w1, w2, theta =  0.5, 0.5, 0.7
  tmp = w1*x1 + w2*x2
  if tmp > theta:
    return 1
  else:
    return 0
```

```python
AND_gate(1, 0)
```
