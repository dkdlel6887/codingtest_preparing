```python
import numpy as np
import matplotlib.pyplot as plt
```

##미분 이용해 접선 그리기

```python
# f(x) = 3x^2 + 4x - 5 , x=1
# 함수 만들기
def my_func(x):
    return 3*x**2 + 4*x -5    #위의 함수식 표현
```

```python
def my_func_diff(x):
    return 6*x + 4    # 미분
```

- y - f(a) = f'(a)(x-a)<br>
- y = f'(a)x - af'(a) + f(a)

```python
x = np.linspace(-10,10)
y = my_func(x)
plt.plot(x,y)
plt.show()
```

```python
x = np.linspace(-10,10)
a=1
y_t = my_func_diff(a)*x - a*my_func_diff(a) + my_func(a)  # 접선의 방정식
plt.plot(x, y)
plt.plot(x,y_t)
plt.show()
```
