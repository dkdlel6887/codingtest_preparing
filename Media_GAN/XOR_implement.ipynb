```python
def AND(x1,x2):
    if x1+x2 > 1:
        return 1
    else: return 0

def OR(x1, x2):
    if x1+x2 > 0:
        return 1
    else: return 0

def NAND(x1,x2):
    if x1+x2 != 2:
        return 1
    else: return 0

def XOR(x1, x2):
    return AND(NAND(x1,x2),OR(x1,x2))
```

```python
print(XOR(0,0))
print(XOR(1,0))
print(XOR(0,1))
print(XOR(1,1))
```
